```python
import csv
import json
import os
import numpy as np
import pandas as pd
import requests
```

```python
def get_request(url):
  try:
     response = requests.get(url=url)
  except:
    print("Hata")
  if response:
    return response.json()
  else:
    return "empty"
```

```python
data = get_request('https://steamspy.com/api.php?request=all&page=0')

state = True
i = 0
while state:
  i +=1
  url = 'https://steamspy.com/api.php?request=all&page=' + str(i)
  data2 = get_request(url)
  
  if data2 != "empty":
    data.update(data2)
  else:
    state = False

# data2 = get_request('https://steamspy.com/api.php?request=tag&tag=Early+Access')
# if data2 != "empty":
#     data.update(data2)
# data2 = get_request('https://steamspy.com/api.php?request=genre&genre=Early+Access')
# if data2 != "empty":
#     data.update(data2)
data2 = get_request('https://steamspy.com/api.php?request=top100in2weeks')
if data2 != "empty":
    data.update(data2)
data2 = get_request('https://steamspy.com/api.php?request=top100forever')
if data2 != "empty":
    data.update(data2)
data2 = get_request('https://steamspy.com/api.php?request=top100owned')
if data2 != "empty":
    data.update(data2)


df = pd.DataFrame.from_dict(data, orient='index')
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
dataset = []
k = 0
for i in df['appid'].unique():
  k += 1
  url = 'https://steamspy.com/api.php?request=appdetails&appid=' + str(i)
  r = requests.get(url)
  soup = BeautifulSoup(r.content, "lxml")
  st = soup.find("p").text

  tags = (st.split('tags":')[1])[:-1]
  st = (st.split('tags":')[0])[:-2]
  appid = st.split('"appid":')[1].split(',')[0]
  name = st.split('"name":"')[1].split('",')[0]
  developer = st.split('"developer":"')[1].split('",')[0]
  publisher = st.split('"publisher":"')[1].split('",')[0]
  positive = st.split('"positive":')[1].split(',')[0]
  negative = st.split('"negative":')[1].split(',')[0]
  average_forever = st.split('"average_forever":')[1].split(',')[0]
  average_2weeks = st.split('"average_2weeks":')[1].split(',')[0]
  median_forever = st.split('"median_forever":')[1].split(',')[0]
  median_2weeks = st.split('"median_2weeks":')[1].split(',')[0]
  ccu = st.split('"ccu":')[1].split(',')[0]
  genre = st.split('"genre":"')[1].split('"')[0]

  dataset.append([appid, name, developer, publisher, positive, negative, average_forever, average_2weeks, median_forever, median_2weeks, ccu, genre, tags])
  if k%50 == 0:
    print(k)
```

```python
df2 = pd.DataFrame(dataset)
```

```python
import pandas as pd
import numpy as np

path = "/GameRecommender/MyDrive/GameRecommender/steamDataSet.csv"
df2 = pd.read_csv(path, sep="|")
df2
```

```python
import re
def editTags(tags):
  return re.sub('[0-9]', '', tags).replace(":,","|").replace('"','').replace('{','').replace('}','').replace(':','').replace(' ', '-').replace('|',' ')
```

```python
df2['editedTags'] = df2['tags'].apply(editTags)
```

```python
df2.head()
```

```python
(df2['editedTags'] + " " + df2['publisher']).iloc[0]
```

```python
df2['editedTags'] = df2['editedTags'] + " " + df2['publisher']
```

```python
new_df = df2[['game', 'editedTags']]
```

```python
new_df['editedTags'] = new_df['editedTags'].apply(lambda X:X.lower())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
vectors = tfidf.fit_transform(new_df['editedTags'])
```

```python
vectors.shape
```

```python
import nltk
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])[1:6]
```

```python
def recommend_game(game):
  game_index = new_df[new_df['game'] == game].index[0]
  distances = similarity[game_index]
  game_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]

  for i in game_list:
    print(new_df.iloc[i[0]].game)
    
```
